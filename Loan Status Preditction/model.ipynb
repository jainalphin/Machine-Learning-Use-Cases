```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df=pd.read_csv('train.csv')
df
```

```python
df.columns =df.columns.str.lower()
```

```python
df['dependents'].value_counts()
```

```python
df.isna().sum()
```

```python
df2=df.copy()
```

```python
df2['gender']=df2['gender'].fillna(df2['gender'].mode()[0])
df2['married']=df2['married'].fillna(df2['married'].mode()[0])
df2['dependents']=df2['dependents'].fillna(df2['dependents'].mode()[0])
df2['self_employed']=df2['self_employed'].fillna(df2['self_employed'].mode()[0])
df2['loan_amount_term']=df2['loan_amount_term'].fillna(df2['loan_amount_term'].median())
df2['loanamount']=df2['loanamount'].fillna(df2['loanamount'].median())
df2['credit_history']=df2['credit_history'].fillna(df2['credit_history'].median())
```

```python
df2.isna().sum()
```

```python
df2.head()
```

```python
df2.drop('loan_id',axis=1,inplace=True)
```

```python
df3=df2.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder =LabelEncoder()
for i in ['gender','dependents','married','education','self_employed','property_area','loan_status']:
    df3[i]=encoder.fit_transform(df3[i])
```

```python
df3
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
scaled_data = pd.DataFrame(minmax_scaler.fit_transform(df3),columns=df3.columns)
scaled_data
```

```python
X=scaled_data.drop('loan_status',axis=1)
y=scaled_data['loan_status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
sns.pairplot(df2)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
logestic_regression=LogisticRegression()
logestic_regression.fit(X_train,y_train)
y_pred=logestic_regression.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest=RandomForestClassifier()
random_forest.fit(X_train,y_train)
y_pred=random_forest.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=22)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.svm import SVC
scv=SVC()
scv.fit(X_train,y_train)
y_pred=scv.predict(X_test)
accuracy_score(y_test,y_pred)
```

# HyperParameter

```python
# Random search Cv

#number of tree in Random Searcv
n_estimators =[int(i) for i in np.linspace(start=100,stop=1200,num=12)]

#Number of fearures to consider at every split
max_fearures=['auto','sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]

# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]

# Method of selecting samples for training each tree
# bootstrap = [True, False]
```

```python
random_grid ={
    'n_estimators':n_estimators,
    'max_features':max_fearures,
    'max_depth':max_depth,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestClassifier()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, 
                               param_distributions = random_grid,
                               scoring='accuracy', 
                               n_iter = 10, 
                               cv = 5, 
                               verbose=2,
                               random_state=42, 
                               n_jobs = 1)
```

```python
rf_random.fit(X,y)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```
