```python
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#here we dont have setnances bcoz it work like that
#so we have to make them to make it as char by char
vectorizer=TfidfVectorizer(tokenizer= word_divide_char)
X=vectorizer.fit_transform(X)
```

## Logestic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
#Logestic regrression is for bianry but some changes i can use for multiclass classifier
log_classification =LogisticRegression(penalty='l2',multi_class='ovr')
```

```python
log_classification.fit(X_train,y_train)
```

```python
print(log_classification.score(X_test,y_test))
```

```python
log_classification2 =LogisticRegression(solver='newton-cg',multi_class='multinomial',random_state=0)
log_classification2.fit(X_train,y_train)
print(log_classification2.score(X_test,y_test))
```

```python
X_predict =np.array(['E!imqCz$^E95!mE^V2X5tB&Bd'])
X_predict=vectorizer.transform(X_predict)
y_pred=log_classification2.predict(X_predict)
print(y_pred)
```

```python
import pickle
#store the content
with open("transformer.pkl", 'wb') as transformer:
    pickle.dump(vectorizer, transformer)
```
